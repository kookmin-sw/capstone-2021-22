```
# 데이터셋 불러오기
data_aug_gen = ImageDataGenerator(rescale=1./255, 
                                  rotation_range=15,
                                  width_shift_range=0.1,
                                  height_shift_range=0.1,
                                  shear_range=0.5,
                                  zoom_range=[0.8, 2.0],
                                  horizontal_flip=True,
                                  vertical_flip=True,
                                  fill_mode='nearest')

file_list = os.listdir("./sampleimg")
file_list_sub = [0 for i in range(len(file_list))]
for j in range (0, len(file_list)):
  file_list_sub[j] = file_list[j].rstrip('.png')
print(file_list_sub)
print(len(file_list))
```

```
name_idx = 0
for file in file_list:
  fname = "./sampleimg/%s" %file
  img = load_img(fname)
  x = img_to_array(img)
  x = x.reshape((1,) + x.shape)

  i = 0
  for batch in data_aug_gen.flow(x, batch_size=1, save_to_dir='./sampleimg_nobackground', save_prefix=file_list_sub[name_idx], save_format='jpg'):
      i += 1
      print(file_list[name_idx],i)
      if i >99:
          break
  name_idx = name_idx+1


# fname = "/content/drive/MyDrive/sampleimg/195900001.jpg"
# img = load_img(fname)
# x = img_to_array(img)
# x = x.reshape((1,) + x.shape)

# i = 0
# for batch in data_aug_gen.flow(x, batch_size=1, save_to_dir='/content/drive/MyDrive/sampleimg/195900001', save_prefix=i, save_format='jpg'):
#     i += 1
#     print(i)
#     if i >29: 
#         break
```
