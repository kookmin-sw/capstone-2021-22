```
import numpy as np
import os
import pandas as pd
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
np.random.seed(5)

data = pd.read_excel('/content/drive/MyDrive/이게뭐약/공공데이터개방_낱알식별목록.xlsx')
for i in range(len(data['품목명'])):
  data['품목명'][i] = data['품목명'][i].split('(')[0]
  data['품목명'][i] = data['품목명'][i].replace('/','-')
```

```
# 데이터셋 불러오기
data_aug_gen = ImageDataGenerator(rescale=1./255, 
                                  rotation_range=15,
                                  width_shift_range=0.1,
                                  height_shift_range=0.1,
                                  shear_range=0.5,
                                  zoom_range=[0.8, 2.0],
                                  horizontal_flip=True,
                                  vertical_flip=True,
                                  fill_mode='nearest')

file_list = os.listdir("/content/drive/MyDrive/이게뭐약/sampleimgdata/sampling_remove_background")
# print(file_list)
# print(len(file_list))
```

```
name_idx = 0
for file in file_list:
  fname = "/content/drive/MyDrive/이게뭐약/sampleimgdata/sampling_remove_background/%s" %file
  img = load_img(fname)
  x = img_to_array(img)
  x = x.reshape((1,) + x.shape)

  i = 0
  for batch in data_aug_gen.flow(x, batch_size=1, save_to_dir='/content/drive/MyDrive/이게뭐약/sampleimgdata/sampleimg_파일명으로class구분', save_prefix=data['품목명'][name_idx], save_format='jpg'):
      i += 1
      print(data['품목명'][name_idx],i)
      if i >29:
          break
  name_idx = name_idx+1


# fname = "/content/drive/MyDrive/sampleimg/195900001.jpg"
# img = load_img(fname)
# x = img_to_array(img)
# x = x.reshape((1,) + x.shape)

# i = 0
# for batch in data_aug_gen.flow(x, batch_size=1, save_to_dir='/content/drive/MyDrive/sampleimg/195900001', save_prefix=i, save_format='jpg'):
#     i += 1
#     print(i)
#     if i >29: 
#         break
```
