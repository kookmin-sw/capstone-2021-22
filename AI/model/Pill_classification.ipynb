```python
image_dir = '/content/sampleimg_nobackground'
print(image_dir)
```

```python
image_list = os.listdir(image_dir)
print(image_list)
```

```python
#png 파일 있으면 jpg 로 변환해주기
from PIL import Image

for sampleimg in image_list :
  if os.path.splitext(sampleimg)[-1] == '.png':
    im = Image.open(image_dir + '/'+sampleimg)
    im = im.convert("RGB")
    im.save(image_dir + '/'+'%s.jpg'%os.path.splitext(sampleimg)[0])
    print(sampleimg)
    os.remove(image_dir + '/'+sampleimg)
```

```python
#파일명 유니코드 변환해주기
import unicodedata
for imgname in image_list :
  sampleimg2 = unicodedata.normalize('NFC',imgname)
  os.rename(image_dir+'/'+imgname, image_dir+'/'+sampleimg2)
  
```

```python
#전체 이미지수
allimage = [f for f in image_list]
print(len(allimage))
```

```python
classlist = []
for image in allimage :
  classname = re.split('_|.jpg',image)[0]
  if classname not in classlist:
    classlist.append(classname)

print(len(classlist))
# # classlist.sort()
listt = sorted(sorted(classlist), key=lambda c:0 if re.search('[ㄱ-힣]', c) else 1)
listt
```
